```python
#check version
import pycaret
print(pycaret.__version__)
```

```python
#loading sample data from pycaret dataset module
from pycaret.datasets import get_data
data = get_data('diamond')
```

```python
#import pycaret anomaly and init setup
from pycaret.anomaly import *
s = setup(data, session_id = 123)
```

```python
#import AnomalyExperiment and init the class
from pycaret.anomaly import AnomalyExperiment
exp = AnomalyExperiment()
```

```python
#type of exp
type(exp)
```

```python
#init setup on exp
exp.setup(data, session_id = 123)
```

```python
#Create Model
#Train iforest Model
iforest = create_model('iforest')
iforest
```

```python
#Check all available models
models()
```

```python
#Assign Model
iforest_anomalies = assign_model(iforest)
iforest_anomalies
```

```python
#Analyze Model
#tsne plot anomalies
plot_model(iforest, plot='tsne')
```

```python
#check docstring for available plots
help(plot_model)
```

```python
#Alternative
evaluate_model(iforest)
```

```python
#Prediction
iforest_pred = predict_model(iforest, data=data)
iforest_pred
```

```python
#SaveModel
save_model(iforest, 'iforest_pipeline')
```
