```python
# create a custom function
import numpy as np

def custom_metric(y, y_pred):
  tp = np.where((y == 1) & (y_pred == 1), (100), 0)
  fp = np.where((y == 0) & (y_pred == 1), -5, 0)
  return np.sum([tp,fp])

  #add metric to PyCaret
  add_metric('custom_metric', 'Custom Metric', custom_metric)
```
