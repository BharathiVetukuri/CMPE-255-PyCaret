```python
from pycaret.datasets import get_data
data = get_data('germany')
```

```python
mb = data.groupby(['InvoiceNo', 'Description'])['Quantity'].sum().unstack().fillna(0)
```

```python
mb = mb.applymap(lambda x: 1 if x > 0 else 0)
```

```python
print(mb.head())
```

```python
frequent_itemsets = apriori(mb, min_support=0.07, use_colnames=True)
print(frequent_itemsets.head())
```

```python
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1)
print(rules.head())
```

```python
import matplotlib.pyplot as plt
import networkx as nx
import seaborn as sns
```

```python
#Scatter plot of support vs. confidence, with lift as color scale
plt.figure(figsize=(12, 5))
sns.scatterplot(x='support', y='confidence', hue='lift', data=rules, palette='viridis', size='lift', alpha=0.6)
plt.title('Association Rules: Support vs. Confidence')
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.show()
```

```python
#3D Scatter Plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

ax.set_xlabel('Support')
ax.set_ylabel('Confidence')
ax.set_zlabel('Lift')

ax.scatter(rules['support'], rules['confidence'], rules['lift'], c=rules['lift'], cmap='viridis')

ax.set_title('3D Plot of Association Rules: Support, Confidence and Lift')
plt.show()
```

```python
#HeatMap of Lift Values
rules_pivot = rules.pivot(index='antecedents', columns='consequents', values='lift')
plt.figure(figsize=(10, 8))
sns.heatmap(rules_pivot, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap of Lift Values')
plt.xlabel('Consequents')
plt.ylabel('Antecedents')
plt.show()
```
